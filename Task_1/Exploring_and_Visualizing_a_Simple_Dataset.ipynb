```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
```

```python
# Display basic information
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nFirst 5 Rows:")
df.head()
```

```python
# --- Scatter Plot ---
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species')
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

```python
# --- Histogram ---
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='petal length (cm)', kde=True, bins=15)
plt.title("Histogram: Distribution of Petal Length")
plt.show()
```

```python
# --- Box Plot ---
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='species', y='petal width (cm)')
plt.title("Box Plot: Petal Width by Species")
plt.show()
```

‚úÖ **Summary:**  
Through these visualizations, we explored how iris flower measurements differ across species.  
- The **scatter plot** shows variable relationships.  
- The **histogram** highlights distribution patterns.  
- The **box plot** helps spot outliers and compare feature ranges.
